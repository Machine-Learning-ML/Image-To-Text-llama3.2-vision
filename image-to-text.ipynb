```python
import cv2
import pytesseract
from PIL import Image

# 1. Load the image
image = cv2.imread('/home/kosal/AI/studycircle-ai/image/handnote-11.png')   # Replace with your image path

# 2. (Optional) Preprocess the image for better OCR results
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# 3. Set Tesseract path if needbed (especially for Windows)
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# 4. OCR: Extract text
text = pytesseract.image_to_string(thresh)
print("Extracted Text:")
print(text)
```
